```python
import requests
import pprint
from getpass import getpass
```

```python
api_key = getpass('APIキー')
```

```python
search_adress = input('最寄駅')
```

```python
url = 'https://maps.googleapis.com/maps/api/geocode/json'
response = requests.get(
    url,
    params={
        'address' : search_adress,
        'key' : api_key
    }
)
```

```python
json_data_adress = response.json()
pprint.pprint(json_data_adress)
```

```python
lat = json_data_adress['results'][0]['geometry']['location']['lat']
lon = json_data_adress['results'][0]['geometry']['location']['lng']

print(lat, lon)
```

```python
url1 = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'
response1 = requests.get(
    url1,
    params={
        'key' : api_key,
        'location': f'{lat},{lon}',
        'language' : 'ja',
        'radius' : 500,
        'type' : 'restaurant',
    }
)
```

```python
json_data_restaurant = response1.json()
pprint.pprint(json_data_restaurant)
```

```python
for results in json_data_restaurant['results']:
  name = results['name']
  rating = results['rating']
  place = results['vicinity']
  print(name, rating, place)
```
